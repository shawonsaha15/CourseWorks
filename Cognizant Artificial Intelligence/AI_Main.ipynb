```python
def convert_to_datetime(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format='%Y-%m-%d %H:%M:%S')
  return dummy
```

```python
sales_df = convert_to_datetime(sales_df, 'timestamp')
stock_df = convert_to_datetime(stock_df, 'timestamp')
temp_df = convert_to_datetime(temp_df, 'timestamp')

print("Sales:")
sales_df.info()
print("\nStock:")
stock_df.info()
print("\nTemperature:")
temp_df.info()
```

## Section 4 - Data merging

Here we will merge the tables into one.

```python
from datetime import datetime

def convert_timestamp_to_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy
```

```python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
stock_df = convert_timestamp_to_hourly(stock_df, 'timestamp')
temp_df = convert_timestamp_to_hourly(temp_df, 'timestamp')

sales_df.head()
```

For sales data we will group the the columns `timestamp` and `product_id` and aggregate quantity.

```python
sales_agg = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_agg.head()
```

```python
stock_agg = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_agg.head()
```

```python
temp_agg = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_agg.head()
```

```python
merged_df = stock_agg.merge(sales_agg, on=['timestamp', 'product_id'], how='left')
merged_df.head()
```

```python
merged_df = merged_df.merge(temp_agg, on=['timestamp'], how='left')
merged_df.head()
```

```python
merged_df.isnull().sum()
```

As we can see that there are null values present, we will fill the null values with 0.

```python
merged_df['quantity'] = merged_df['quantity'].fillna(0)
merged_df.info()
```

Combining more features to the merged table such as `categories` and `unit_price`

```python
product_categories = sales_df[['product_id', 'category']].drop_duplicates()
product_price = sales_df[['product_id', 'unit_price']].drop_duplicates()
```

```python
merged_df = merged_df.merge(product_categories, on=['product_id'], how='left')
merged_df = merged_df.merge(product_price, on=['product_id'], how='left')
merged_df.head()
```

## Section 5 - Feature engineering

We have our cleaned and merged data and now we must transform this data so that the columns are in a suitable format for a machine learning model.

```python
merged_df['timestamp_day_of_month'] = merged_df['timestamp'].dt.day
merged_df['timestamp_day_of_week'] = merged_df['timestamp'].dt.dayofweek
merged_df['timestamp_hour'] = merged_df['timestamp'].dt.hour
merged_df.drop(columns=['timestamp'], inplace=True)
merged_df.head()
```

Next, we will do one hot encoding on the `category` column.

```python
merged_df = pd.get_dummies(merged_df, columns=['category'])
merged_df.head()
```

```python
merged_df.info()
```

As we can see, there is only one feature that has object type data. Therefore, we will drop it as it can no longer be further converted to numeric data.

```python
merged_df.drop(columns=['product_id'], inplace=True)
merged_df.head()
```

### Section 6 - Data modelling

Here, we will create the machine learning models to run the cleaned dataset.

```python
X = merged_df.drop(columns=['estimated_stock_pct'])
y = merged_df['estimated_stock_pct']

print(X.shape)
print(y.shape)
```

Splitting the data into 80% training and 20% testing sets. We will run the Random Forest regressor for 10 folds and calculate the mean absolute error for each folds.

```python
K = 10
split = 0.20
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
accuracy = []

for fold in range(0, K):

  # Instantiate algorithm
  model = RandomForestRegressor()
  scaler = StandardScaler()

  # Create train and test samples
  X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=split, random_state=42)

  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test = scaler.transform(X_test)

  # Train model
  trained_model = model.fit(X_train, y_train)

  # Generate predictions on test sample
  y_pred = trained_model.predict(X_test)

  # Compute accuracy, using mean absolute error
  mae = mean_absolute_error(y_test, y_pred)
  accuracy.append(mae)
  print(f"Fold {fold + 1}: MAE = {mae:.3f}")

print(f"Average MAE: {(sum(accuracy) / len(accuracy)):.2f}")
```

We can see that `mean absolute error` (MAE) is exactly same each time. This shows that the performance is consistent across different random samples of the data.

MAE was chosen as the evaluation metric because it describes how closely the machine learning model was able to predict the exact value of `estimated_stock_pct`.

We have analysed that even though the model is working fine, the MAE results are not so good, since the average value of the target variable is high. In ideal world, we would want the MAE to be as low as possible.

To improve the results, we need more findings to add more dataset and have it further engineered. To solidify our analyzation, we shall add visualizations using `matplotlib` and `numpy` libraries.

```python
import numpy as np
import matplotlib.pyplot as plt

features = [i.split("__")[0] for i in X.columns]
importances = model.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(10, 10))
plt.title('Feature Importance')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

The feature importance visualisation tells us that:
* The categories do not have much effect
* The `unit_price` and `temperature` were the most important in predicting stock
* The hour of the day was also important
